```python
#W7A1
def binary_search(arr, left, right, target):
    while left <= right:
        mid = (left + right) // 2

        if arr[mid] == target:
            return mid
        elif target < arr[mid]:
            right = mid - 1
        else:
            left = mid + 1

    return -1


m = list(map(int, input().split()))
x = int(input())

print(binary_search(m, 0, len(m) - 1, x))

```

```python
#W7A3(Bubble sort)
def bubble_sort(x):
    n = len(x)
    for i in range(0,n-1):
        for j in range(0,n-1-i):
            if x[j] > x[j+1]:
                x[j] ,x[j+1] = x[j+1],x[j]
    return x


a = list(map(int,input().split()))
print(bubble_sort(a))
```

```python
#W7A3(Selection sort)
def selection_sort(x) :
    n = len(x)
    for i in range(0,n-1):
        min_index = i 
        for j in range(i+1,n) :
            if x[i] > x[j] :
                min_index = j
        x[i], x[min_index] = x[min_index] , x[i]
    return x
a = list(map(int,input().split()))
print(selection_sort(a))
```

```python
#W7A5

def value(x,y):
    n = len(x)
    tong = 0
    for a in range(0,n-2):
        for b in range(a+1,n-1):
            for c in range (b+1,n):
                if x[a]+x[b]-y*x[c] ==0 or x[a]+x[c]-y*x[b] == 0 or x[b]+x[c]-y*x[a] == 0:
                    tong += 1
    return tong

m = list(map(int,input().split()))
n = int(input())
print(value(m,n))


```

```python
#W7A7(Với trường hợp số quá lớn nhưng dãy cũng quá xa(Sai))
nums = list(map(int,input().split()))

n = len(nums)
k = [nums[0]]

for i in range (1,n):
    m = len(k)
    if nums[i] > k[m-1]:
        k.append(nums[i])
    if nums[i] < k[m-1]:
        k.remove(k[m-1])
        k.append(nums[i])
print(len(k))
    
```
