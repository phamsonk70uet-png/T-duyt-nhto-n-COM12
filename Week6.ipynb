```python
#W6A19
def max_index(a):
    max_number = a[0]
    idx = 0
    for i in range (1, len(a)):
        if a[i] > max_number:
            max_number = a[i]
            idx = i
    return idx

a = list(map(int,input().split()))
print(max_index(a))
```
