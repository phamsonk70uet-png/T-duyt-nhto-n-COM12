```python
#W8A1
class Cylinder:
    def __init__(self,bk,ccao):
        self.bk=bk
        self.ccao=ccao
    def dien_tich(self):
        s=2*3.14*(self.bk**2)+2*3.14*self.bk*self.ccao
        return round(s,2)
    def the_tich(self):
        V=3.14*(self.bk**2)*self.ccao
        return round(V,2)
r,h=map(float,input().split())
Cylinder1=Cylinder(r,h)
print(Cylinder1.dien_tich())
print(Cylinder1.the_tich())
```

```python
#W8A2
class Date:
    def __init__(self,ngay,thang,nam):
        self.ngay=ngay
        self.thang=thang
        self.nam=nam
    def nam_nhuan(self):
        return self.nam%400==0 or (self.nam%4==0 or self.nam%100!=0)
    def so_ngay_trong_thang(self):
        if self.thang in [1,3,5,7,8,10,12]:
            return 31
        elif self.thang in [2,4,6,9,11]:
            return 30
        elif self.thang==2:
            return 29 if self.nam_nhuan() else 28
        else:
            return 0
    def hop_le(self):
        if self.nam<1 or self.thang<1 or self.thang>12:
            return False
        if self.ngay<1 or self.ngay>self.so_ngay_trong_thang():
            return False
        return True
    def ngay_ke_tiep(self):
        if not self.hop_le():
            return None
        ngay=self.ngay+1
        thang=self.thang
        nam=self.nam
        if ngay>self.so_ngay_trong_thang():
            ngay=1
            thang+=1
            if thang>12:
                thang=1
                nam+=1
        return Date(ngay,thang,nam)

n=input().strip()
try:
    a,b,c=map(int,n.split("/"))
    ngay=Date(a,b,c)
    if not ngay.hop_le():
        print("INVALID")
    else:
        next_day=ngay.ngay_ke_tiep()
        print(f"{next_day.ngay}/{next_day.thang}/{next_day.nam}")
except:
    print("INVALID")
```

```python
#W8A3
import math
class Point:
    def __init__(self,x,y):
        self.x=x
        self.y=y
    def goc_toa_do(self):
        return self.x==0 and self.y==0
    def tren_truc_tung(self):
        return self.x==0 and self.y!=0
    def tren_truc_hoanh(self):
        return self.x!=0 and self.y==0
    def khoang_cach(self,goc):
        return math.sqrt((self.x-goc.x)**2+(self.y-goc.y)**2)
x,y=map(float,input().split())
A=Point(x,y)
O=Point(0,0)
if A.goc_toa_do():
    print("A tren goc toa do")
else:
    print("A khong la tren goc toa do")
if A.tren_truc_tung():
    print("A nam tren truc tung")
else:
    print("A khong nam tren truc tung")
if A.tren_truc_hoanh():
    print("A nam tren truc hoanh")
else:
    print("A khong nam tren truc hoanh")

kc=A.khoang_cach(O)
print(f"Khoang cach tu A den goc toa do la: {kc:.2f}")
```

```python
#W8A6
class Rectangle:
    def __init__(self,w,h):
        self.w=w
        self.h=h
    def area(self):
        return self.w*self.h
    def perimeter(self):
        return 2*(self.w+self.h)
    def scale(self,k):
        self.w=self.w*k
        self.h=self.h*k
w,h,k=map(float,input().split())
a=Rectangle(w,h)
a.scale(k)
area=round(a.area(),2)
perimeter=round(a.perimeter(),2)
print(f"{area:.2f} {perimeter:.2f}")
```

```python
# W8A4: Shopping Cart
class ShoppingCart:
    def __init__(self): self.items = []
    def add_item(self, name, price): self.items.append({'name': name, 'price': price})
    def remove_item(self, name):
        self.items = [i for i in self.items if i['name'] != name]
    def is_empty(self): return len(self.items) == 0
    def total(self): return sum(i['price'] for i in self.items)
    def show(self): print(self.items)
    def clear(self): self.items = []
```

```python
#W8A8
class BankAccount:
    def __init__(self, o, b): self.o, self.b = o, b
    def deposit(self, x): self.b += x
    def withdraw(self, x): self.b -= x if x <= self.b else 0
    def __str__(self): return f"{self.o}: {self.b:.2f}"
```

```python
#W8A9
class Student:
    def __init__(self, n): self.n, self.s = n, {}
    def add_score(self, subj, scr): self.s[subj] = scr
    def avg(self): return sum(self.s.values()) / len(self.s) if self.s else 0
    def rank(self):
        a = self.avg()
        return "Excellent" if a >= 8 else "Good" if a >= 6.5 else "Average" if a >= 5 else "Poor"
```

```python
#W8A11
import math
def angle_3d(A, B, C, D):
    vec = lambda p1, p2: [c2-c1 for c1, c2 in zip(p1, p2)]
    cross = lambda u, v: [u[1]*v[2]-u[2]*v[1], u[2]*v[0]-u[0]*v[2], u[0]*v[1]-u[1]*v[0]]
    mag = lambda v: math.sqrt(sum(x**2 for x in v))
    n1, n2 = cross(vec(B, A), vec(B, C)), cross(vec(C, B), vec(C, D))
    cos_t = sum(a*b for a, b in zip(n1, n2)) / (mag(n1) * mag(n2))
    return round(math.degrees(math.acos(min(abs(cos_t), 1.0))), 2)
```

```python
class ComplexNumber:
    def __init__(self, r, i): self.r, self.i = r, i
    def multiply(self, o): 
        return ComplexNumber(self.r*o.r - self.i*o.i, self.r*o.i + self.i*o.r)
    def __str__(self): return f"{self.r} {self.i}"
```

```python
#W8A13
class Triangle:
    def __init__(self,x,y,z):
        self.x =x
        self.y =y
        self.z =z
    def getArea(self):
        a, b, c = self.x, self.y, self.z
        if a <= 0 or b <= 0 or c <= 0:
            return -1
        if a + b <= c or a + c <= b or b + c <= a:
            return -1
        p = (a + b + c) / 2
        return ((p * (p - a) * (p - b) * (p - c)))**0.5
a, b, c = map(int, input().split())
triangle = Triangle(a, b, c)
area = triangle.getArea()
if area == -1:
    print("invalid")
else:
    print(f"{area:.2f}")
```
