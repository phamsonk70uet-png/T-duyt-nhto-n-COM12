```python
#W4A6
def check_so_nguyen_to (n):
    if n < 2:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True
a, b = map(int, input().split())
c = 0
for i in range(a, b+1):
    if check_so_nguyen_to(i):
        c += i
print(c)
```

```python
#w4A9
def square_num(n):
    x = 1
    while x ** 2 < n:
        x += 1
    return x ** 2 == n
n = int(input())
for i in range(1, n):
    if square_num(i):
        print(i, end =' ')
```

```python
#W4A13
def uoc(n):
    sum = 0
    for i in range(1, n):
        if n % i == 0:
            sum += i
    return sum
a, b = map(int, input().split())
if uoc(a) == b and uoc(b) == a:
    print("True")
else:
    print("False")
```
