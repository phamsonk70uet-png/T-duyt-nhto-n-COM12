```python
#W9A1
def merge(a, b):
    i = j = 0
    res = []
    while i < len(a) and j < len(b):
        if a[i] < b[j]:
            res.append(a[i])
            i += 1
        elif a[i] == b[j]:
            res.append(a[i])
            res.append(b[j])
            i += 1
            j += 1
        else:
            res.append(b[j])
            j += 1
    res.extend(a[i:])
    res.extend(b[j:])
    return res
a = list(map(int, input().split()))
b = list(map(int, input().split()))
print(merge(a, b))
```

```python
#W9A2
def find_k(matrix, k):
    n = len(matrix)
    m = len(matrix[0])
    left = matrix[0][0]
    right = matrix[n - 1][m - 1]
    while left < right:
        mid = (left + right) // 2
        count = 0
        i = n - 1
        j = 0
        while i >= 0 and j < m:
            if matrix[i][j] <= mid:
                count += i + 1
                j += 1
            else:
                i -= 1
        if count < k:
            left = mid + 1
        else:
            right = mid
    return left
n, m = map(int, input().split())
matrix = []
for _ in range(n):
    row = list(map(int, input().split()))
    matrix.append(row)
k = int(input())
print(find_k(matrix, k))
```

```python
#W9A3
def m(n):
    s = list(str(n))
    l = [-1] * 10
    for i, c in enumerate(s):
        l[int(c)] = i
    for i, c in enumerate(s):
        for d in range(int(c)):
            if l[d] > i:
                s[i], s[l[d]] = s[l[d]], s[i]
                return int(''.join(s))
    s[-1], s[-2] = s[-2], s[-1]
    return int(''.join(s))
n=input()
print(m(n))
```

```python
#W9A4
def can_transform(a, b):
    n, m = len(a), len(b)
    dp = [[False] * (m + 1) for _ in range(n + 1)]
    dp[0][0] = True
    for i in range(1, n + 1):
        for j in range(m + 1):
            if a[i - 1].islower():
                dp[i][j] |= dp[i - 1][j]
                if j > 0 and a[i - 1].upper() == b[j - 1]:
                    dp[i][j] |= dp[i - 1][j - 1]
            else:
                if j > 0 and a[i - 1] == b[j - 1]:
                    dp[i][j] |= dp[i - 1][j - 1]

    return "YES" if dp[n][m] else "NO"
Q = int(input())
for _ in range(Q):
    a = input().strip()
    b = input().strip()
    print(can_transform(a, b))
```

```python
#W9A7
def canPlaceFlower(f, k):
    if k == 0:
        return True
    count = 0
    if len(f) == 1:
        if f[0] == 0:
            count += 1
    else:    
        if f[0] == 0 and f[1] == 0:
            count += 1
            f[0] = 1
        for i in range(1, len(f) - 1):
            if f[i] == 0 and f[i - 1] == 0 and f[i + 1] == 0:
                count += 1
                f[i] = 1
        if f[len(f) - 1] == 0 and f[len(f) - 2] == 0:
            count += 1
            f[len(f) - 1] = 1
    return count >= k
f = list(map(int, input().split()))
k = int(input())
print(canPlaceFlower(f, k))
```

```python
#W9A8
def is_path_crossing(moves):
    x = 0
    y = 0
    res = {(0,0)}
    for i in range(len(moves)):
        if moves[i] == "R":
            x += 1
        elif moves[i] == "L":
            x -= 1
        elif moves[i] == "U":
            y += 1
        elif moves[i] == "D":
            y -= 1
        if (x,y) in res:
            return True
        res.add((x,y))
    return False
moves = input()
print(is_path_crossing(moves))
```

```python
#W9A9
def Medians(nums):
    minn = []
    maxx = []
    med = []
    median = 0
    if nums[0] < nums[1]:
        minn.append(nums[0])
        maxx.append(nums[1])
    else:
        minn.append(nums[1])
        maxx.append(nums[0])
    if nums[2] < nums[3]:
        minn.append(nums[2])
        maxx.append(nums[3])
    else:
        minn.append(nums[3])
        maxx.append(nums[2])
    med.append(minn[1]) if minn[0] < minn[1] else med.append(minn[0])
    med.append(maxx[1]) if maxx[0] > maxx[1] else med.append(maxx[0])
    med.append(nums[4])
    if med[2] < med[0]:
        median = med[0]
    elif med[2] > med[1]:
        median = med[1]
    else:
        median = med[2]
    return median
a = list(map(int, input().split()))
print(Medians(a))
```
