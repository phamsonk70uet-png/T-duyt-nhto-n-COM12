```python
#W10A1
def divide(a,b):
    try:
        m=a/b
        print(f"{m:.2f}")
    except ZeroDivisionError:
        print("Loi chia khong")

a,b=map(int,input().split())
divide(a,b)
```

```python
#W10A3
def tong_bp(n):
    a=0
    while n!=0:
        b=n%10
        a+=b**2
        n=n//10
    return a
def so_hanhphuc(n):
    a=set()
    while n!=1:
        if n in a:
            return False
        a.add(n)
        n=tong_bp(n)
    return True
try:
    n=int(input())
    if so_hanhphuc(n):
        print("YES")
    else:
        print("NO")
except ValueError as e:
    print(e)
```

```python
#W10A11
t = int(input().strip())
for _ in range(t):
    s = input().strip()
    valid = True
    prev_repeat = False   
    i = 0
    while i < len(s):
        if s[i] == '\\':         
            if i == len(s) - 1:  
                valid = False
                break
            prev_repeat = False
            i += 2
            continue
        if s[i] in '*+':         
            if prev_repeat:      
                valid = False
                break
            prev_repeat = True
        else:
            prev_repeat = False
        i += 1
    print("True" if valid else "False") 
```
